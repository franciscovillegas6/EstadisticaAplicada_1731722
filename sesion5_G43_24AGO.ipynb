```python
##librerías

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## datos
personas=['Juan','Jorge','Ana','Alba']
edades=[23,28,18,31]

#2
dic1 = {'Municipio': ['Escobedo', 'Apodaca', 'Monterrey', 'Guadalupe'], 'Nacimientos':[345,487,372,298],'Muertes':[25,56,39,67],'Divorcios':[28,46,95,19]} #diccionario
datos_mun=pd.DataFrame(dic1)
```

```python
datos_mun
```

```python
#grafica de barras de las edades matplotlib
plt.bar(personas, edades, color='black')
plt.title('Edades de 4 desconocidos')
plt.xlabel('Personas')
plt.ylabel('Edad en años')
plt.show()
```

```python
#gráfica de barras para los nacimientos en mathplot
plt.bar(datos_mun['Municipio'], datos_mun['Nacimientos'], color='c')
plt.title('Nacimientos en N.L')
plt.xlabel('Municipios')
plt.ylabel('Nacimientos')
plt.show()
```

```python
#definimos el pizarrón/lienzo
#normalmente le llamamos fig,
fig,(panel1,panel2)=plt.subplots(1,2)
```

```python
#definimos el pizarrón/lienzo
#normalmente le llamamos fig,
fig,(panel1,panel2)=plt.subplots(2,1)
#grafica 1
fig,panel1.bar(datos_mun['Municipio'], datos_mun['Nacimientos'], color='indigo')

#grafica 2
panel2.bar(personas, edades, color='goldenrod')
```
