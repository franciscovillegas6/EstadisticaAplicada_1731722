# Problema 1

```python
import numpy as np
import pandas as pd
```

```python
peliculas= pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/peliculas.csv')
```

```python
## Inciso   A

#Cantidad de peliculas
peliculas.title.count()

#cantidad de géneros
peliculas_categorias=peliculas.groupby('category').count()
peliculas_categorias
peliculas_categorias.count()
# promedio de duracion de las peliculas
peliculas.length.mean()

print('Existen ',peliculas.title.count(),'peliculas en catálogo ,',peliculas_categorias.description.count(),' géneros, y duran en promedio las películas',peliculas.length.mean(),' minutos')
```

```python
## Inciso   B
peliculas_dormir= pd.DataFrame(peliculas.query('length>120 & category=="Documentary"'))
peliculas_dormir
```

```python
#inciso C
peliculas_categorias=peliculas.groupby('category')['title'].agg(['count'])
peliculas_categorias=pd.DataFrame(peliculas_categorias)
peliculas_categorias.sort_values('count')
```

```python
#inciso D
duracionPromedioPeliculas=peliculas.groupby('category')['length'].agg(['mean'])
duracionPromedioPeliculas=pd.DataFrame(duracionPromedioPeliculas)
duracionPromedioPeliculas.sort_values('mean', ascending=True).round()
```

```python
#Inciso E
catelogia_duracion=[(peliculas['length']<90),(peliculas['length']>=90) & (peliculas['length']<=150),(peliculas['length']>150)]
tipo_de_duracion=['Corta','Normal','Larga']
peliculas['TipoDuracion']=np.select(catelogia_duracion,tipo_de_duracion)
peliculas['TipoDuracion'].value_counts()
peliculas
```

# Problema 2

```python
ciudades= pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/poblacion_Asgard.csv')
```

```python
ciudades
```

```python
# Inciso A
ciudades=pd.DataFrame(ciudades)
ciudades.sort_values(['Estado','Municipio'], ascending=True)
```

```python
# Inciso B
#cantidad de estados y sus municipios
ciudades_estados=ciudades.groupby('Estado')['Municipio'].count()
ciudades_estados=pd.DataFrame(ciudades_estados)
ciudades_estados
```

```python
# cantidad de municipios del pais
ciudades.Municipio.count()
```

```python
print('Existen en el país', ciudades_estados.count(axis=0).mean(),' Estados cada uno con ',ciudades_estados.Municipio.mean(),'Municipios','y en total existen ',ciudades.Municipio.count(), 'Municipios en el pais')
```

```python
#Inciso C
# Poblacion nacional
print('La poblacion del pais es de',ciudades.Población.sum(),'ciudadanos')
```

```python
# poblacion por estado
poblacionEstado=ciudades.groupby('Estado')['Población'].agg(['sum'])
poblacionEstado=pd.DataFrame(poblacionEstado)
poblacionEstado.sort_values('sum')
```

```python
# Inciso D
MedianaPoblacion=ciudades.Población.median()
categoria_poblacion=[(ciudades['Población']>MedianaPoblacion),(ciudades['Población']<MedianaPoblacion)]
tipo_de_poblacion=['Muy poblados','Poco poblados']
ciudades['TipoPoblacion']=np.select(categoria_poblacion,tipo_de_poblacion)
ciudades['TipoPoblacion'].value_counts()
ciudades
```

```python
# Inciso E
poblacionMunicipio=ciudades.groupby(['Estado','Municipio'])['Población'].agg(['sum'])
poblacionMunicipio=pd.DataFrame(poblacionMunicipio)
#5 municipios mas poblados
poblacionMunicipio.sort_values('sum',ascending=False).head(5)
```

```python
#5 municipios menos poblados
poblacionMunicipio.sort_values('sum',ascending=False).tail(5)
```

```python
#Inciso F
max_MunicipioEstado=ciudades.groupby('Estado')['Población'].idxmax()
ciudades.loc[max_MunicipioEstado]
```
