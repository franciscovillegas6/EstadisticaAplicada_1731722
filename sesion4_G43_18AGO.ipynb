## matplotlib

librería para gráficar en python

```python
import matplotlib.pyplot as plt
```

```python
edades=[25,28,22,26]
personas=['Juan', 'María','Robert','Ana']
```

```python
plt.bar(personas,edades);
#el punto y coma es para eliminar la leyenda
```

```python
plt.bar(personas,edades)
plt.title('edad de 4 personas') #es para insertar el nombre
plt.xlabel('personas')
plt.ylabel('edad en años')
plt.show()
```

```python
plt.bar(personas,edades, color=['skyblue', 'black','gray', 'pink'])
plt.title('edad de 4 personas') #es para insertar el nombre
plt.xlabel('personas')
plt.ylabel('edad en años')
plt.show()
```
